```python
# Necessary Libs

# Libs to vectors
import numpy as np
import pandas as pd

# Libs to graph analysis
import matplotlib.pyplot as plt
import seaborn as sns

# Lib to ignore warnings
import warnings


# turnoff warnings
warnings.filterwarnings('ignore')
```

```python
# Reading data
dataset = pd.read_csv('../input/salary-data/Salary_Data.csv')

# Show data
dataset.head()
```

```python
# Size
dataset.shape
```

```python
# Null data
dataset.isnull().sum()
```

```python
# Graph null data
plt.figure(figsize=(10,5))
sns.heatmap(dataset.isnull(),cbar=False);
```

```python
# describe data
dataset.describe()
```

```python
plt.figure( figsize=(20,6))
plt.title('Salary Analysis')
sns.kdeplot(dataset['Salary'], color='green');
```

```python
plt.figure( figsize=(20,5))
plt.title('Experience Analysis')
sns.kdeplot(dataset['YearsExperience'], color='blue');
```

```python
# Distribution
plt.figure( figsize=(10,5))
sns.boxplot(dataset['Salary']);
```

```python
# Distribution
plt.figure( figsize=(10, 5))
sns.scatterplot( data=dataset, x='Salary', y='YearsExperience')
```

```python
sns.regplot( data=dataset, x='Salary', y='YearsExperience');
```

```python
# Correlation between years of experience and salary
cor = dataset.corr()
plt.figure( figsize=(15,8))
sns.heatmap(cor, annot=True, cbar=False);
```

```python
# Spearman Correlation between years of experience and salary
cor = dataset.corr('spearman')
plt.figure( figsize=(15,8))
sns.heatmap(cor, annot=True, cbar=True);
```
