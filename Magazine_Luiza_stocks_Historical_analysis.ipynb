```python
# Libs to modeling and vectores
import numpy as np
import pandas as pd

# Libraries to imput graph
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go


# Library to ignore warnings
import warnings

# turnoff warnings
warnings.filterwarnings('ignore')
```

```python
# reading data
dataset = pd.read_excel("Vase_004 - Magalu - Sem Resolução.xlsx")
```

```python
# Sharing first 5 lines
dataset.head()
```

```python
# Sharing dataset's dimension
dataset.shape
```

```python
# Sharing quality's information
dataset.info()
```

```python
dataset.describe()
```

```python
# time-series analysis
data = dataset.set_index('Data')

data.head()
```

```python
plt.style.use('seaborn-dark')
plt.figure(figsize=(16,5))
plt.title('Análise das ações da magalu - Fechamento', fontsize=14, loc='left')
plt.plot(data.index, data['Fechamento'])

#graphs labels

plt.xlabel('Periodo da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
# show last five lines
data.tail()
```

```python
# grouping data
data['Fechamento'].rolling(5)
```

```python
# mean's grouping data and plotting graph
media_movel = data['Fechamento'].rolling(5).mean()
media_tendencia = data['Fechamento'].rolling(10).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16,5))
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')

plt.plot(data.index, data['Fechamento'])
plt.plot(data.index, media_movel)
plt.plot(data.index, media_tendencia)

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot(data=data, x='Fechamento');
```

```python
# Boxplot Month 

dataset['Mes'] = dataset['Data'].dt.month

plt.figure( figsize=(16, 5) )
sns.boxplot( data=dataset, x='Mes', y='Fechamento');
```

```python
dataset.groupby(['Mes']).describe()['Fechamento']
```

```python
# 
Grafico = go.Figure(
    data=[
          go.Candlestick(
              x= data.index,
              open = data['Abertura'],
              high = data['Maior'],
              low = data['Maior'],
              close = data['Fechamento'],
          )
    ]
)

Grafico.update_layout( xaxis_rangeslider_visible=False)

Grafico.show()
```
